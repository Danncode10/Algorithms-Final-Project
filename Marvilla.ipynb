### üîç Function: `evenDigitsOnly(n)`

**Goal:**  
Check if all the digits in the given integer `n` are even (i.e., divisible by 2).

---

### üß† Step-by-step Explanation:

1. **Convert the number to a string:**  
   First, we convert `n` into a string so we can iterate through each digit.  
   This makes it easier to treat the digits individually.

2. **Loop through each character (digit):**  
   We will loop through each character of the string. Each character represents a digit in the number.

3. **Convert each character back to an integer:**  
   Once we have a character (e.g., `'2'`), we convert it back into an integer (e.g., `2`) so we can check if it is divisible by 2.

4. **Check if the digit is even:**  
   For each digit, check if it's divisible by 2 (`digit % 2 == 0`).  
   - If it is divisible by 2, the digit is even.  
   - If it isn't divisible by 2 (i.e., the remainder isn't 0), the digit is odd.

5. **Return `False` if any digit is odd:**  
   If we encounter any digit that is not even, we can immediately return `False`.

6. **Return `True` if all digits are even:**  
   If we complete the loop and find that all digits are even, we return `True`.

---

### Example:

- **For `n = 2468`:**

   Convert `2468` to the string `"2468"`.  
   Loop through each character: `'2'`, `'4'`, `'6'`, `'8'`.  
   Each digit is checked, and since all are divisible by 2, the result is `True`.

- **For `n = 1234`:**

   Convert `1234` to the string `"1234"`.  
   Loop through each character: `'1'`, `'2'`, `'3'`, `'4'`.  
   The digit `'1'` is not divisible by 2, so the result is `False`.


````python
def evenDigitsOnly(n):
    # Convert the number to a string
    n_str = str(n)
    
    # Loop through each character (digit) in the string
    for digit in n_str:
        # Convert the character back to an integer
        if int(digit) % 2 != 0:
            return False  # If the digit is odd, return False
    
    # If all digits are even, return True
    return True
````

### üß† Code Explanation:

1. **`def evenDigitsOnly(n):`**  
   This defines a function named `evenDigitsOnly` that takes one parameter `n`, which is the integer whose digits we need to check.

2. **`n_str = str(n)`**  
   We convert the integer `n` into a string, so we can easily loop through each character (digit) in it.

3. **`for digit in n_str:`**  
   We start a loop to iterate over each character in `n_str` (the string version of `n`).  
   In each iteration, `digit` will represent one character from the string, e.g., `'2'`, `'4'`, etc.

4. **`if int(digit) % 2 != 0:`**  
   In this step, we convert the `digit` (which is a string) back to an integer using `int(digit)`.  
   We then check if the digit is odd by checking if `int(digit) % 2 != 0`.  
   If this condition is true, it means the digit is odd.

5. **`return False`**  
   If we encounter an odd digit, we immediately return `False` since we know not all digits are even.

6. **`return True`**  
   If we go through the entire string and all digits are even, we reach this point and return `True`.

---

### 3. `extractEachKth(inputArray, k)`

**Goal:**  
Remove every k-th element from the list.

---

### üß† Step-by-step Explanation:

1. **Loop through the input array using index:**  
   Iterate through the `inputArray` by index to access each element.

2. **Keep all elements except those at indices `k-1`, `2k-1`, `3k-1`, etc.:**  
   Skip elements at positions that are multiples of `k` (adjusted for zero-based indexing).

3. **Collect the remaining elements into a new list:**  
   Append all elements that are not at the skipped indices to a new list.

4. **Return the new list:**  
   After the loop finishes, return the new list that contains the remaining elements.


### üîç Function: `extractEachKth(inputArray, k)`

**Goal:**  
Remove every k-th element from a list. For example, if `k = 3`, it will remove every 3rd element in the array.

---

### üß† Step-by-step Explanation:

1. **Loop through the array using the index:**  
   We need to go through each element of the array, and we can use the index of the element to identify which ones we want to remove.

2. **Identify every k-th element:**  
   The k-th element can be identified by looking at indices: `k-1`, `2k-1`, `3k-1`, etc.  
   These indices represent the positions in the list of every k-th element.  
   For example, if `k = 3`, the elements we remove would be at indices 2, 5, 8, etc. (since index counting starts at 0).

3. **Keep all other elements:**  
   For every element whose index is not a multiple of `k-1`, we keep it in the new list.

4. **Create a new list with the remaining elements:**  
   As we loop through the array, we will collect the elements that are not the k-th elements in a new list.

5. **Return the new list:**  
   After finishing the loop, return the list that contains only the elements that were kept (i.e., all non k-th elements).

---

### Example:

- **For `inputArray = [1, 2, 3, 4, 5, 6, 7, 8]` and `k = 3`:**

   We want to remove elements at indices 2, 5, 8, etc. (i.e., the 3rd, 6th, 9th elements).  
   **Result:** The new array would be `[1, 2, 4, 5, 7, 8]`.

- **For `inputArray = [10, 20, 30, 40, 50]` and `k = 2`:**

   We want to remove elements at indices 1, 3, etc. (i.e., the 2nd, 4th elements).  
   **Result:** The new array would be `[10, 30, 50]`.


````python
def extractEachKth(inputArray, k):
    # Create an empty list to store the result
    result = []
    
    # Loop through the input array using its index
    for i in range(len(inputArray)):
        # If the index is not a multiple of k-1, add the element to the result list
        if (i + 1) % k != 0:
            result.append(inputArray[i])
    
    return result

# Example usage:
inputArray = [1, 2, 3, 4, 5, 6, 7, 8]
k = 3
print(extractEachKth(inputArray, k))  # Output: [1, 2, 4, 5, 7, 8]
````

### üß† Code Explanation:

1. **`def extractEachKth(inputArray, k):`**  
   We define the function `extractEachKth` with two parameters:  
   - `inputArray`: The list of elements from which we will remove every k-th element.  
   - `k`: The interval (or "k-th" position) that tells us which elements to remove.

2. **`result = []`**  
   We create an empty list called `result` that will hold the elements after removing every k-th element.

3. **`for i in range(len(inputArray)):`**  
   We loop through each index `i` of the `inputArray`.  
   - `len(inputArray)` gives the total number of elements in the list.  
   - This loop allows us to access both the index and the element at that index.

4. **`if (i + 1) % k != 0:`**  
   We check if the element at index `i` is **not** a multiple of `k`.  
   - `(i + 1)` is used to convert the index from zero-based to one-based numbering.  
   - If `(i + 1) % k != 0`, the element is **not** a k-th element, so we will **keep it**.

5. **`result.append(inputArray[i])`**  
   If the condition is met, we append the element `inputArray[i]` to the `result` list.

6. **`return result`**  
   After the loop completes, we return the `result` list, which contains all elements except every k-th one.

---

### üîç Example:

**For `inputArray = [1, 2, 3, 4, 5, 6, 7, 8]` and `k = 3`:**

We loop through each index:

- Index 0 ‚Üí Not 3rd ‚Üí keep `1`  
- Index 1 ‚Üí Not 3rd ‚Üí keep `2`  
- Index 2 ‚Üí 3rd ‚Üí remove `3`  
- Index 3 ‚Üí Not 3rd ‚Üí keep `4`  
- Index 4 ‚Üí Not 3rd ‚Üí keep `5`  
- Index 5 ‚Üí 6th ‚Üí remove `6`  
- Index 6 ‚Üí Not 3rd ‚Üí keep `7`  
- Index 7 ‚Üí Not 3rd ‚Üí keep `8`  

**Result:** `[1, 2, 4, 5, 7, 8]`

### 5. `factorializeANumber(n)`

#### Goal:
Return the factorial of a number `n`.

#### Steps:
1. Set a variable `result = 1`.
2. Loop from `1` to `n` (inclusive).
3. Multiply `result` by each number in the loop.
4. After the loop ends, return `result`.




### üéØ Goal:
The goal is to calculate the factorial of a number `n`, which is represented as `n!`.  
The factorial of a number is the product of all positive integers less than or equal to that number.

#### Examples:
- `5! = 5 * 4 * 3 * 2 * 1 = 120`
- `4! = 4 * 3 * 2 * 1 = 24`
- `0! = 1` (by definition)

---

### üß† Step-by-Step Explanation:

#### 1. Set a variable `result = 1`
We initialize `result` to 1 because multiplying by 1 has no effect, and it serves as the base for accumulation.

#### 2. Loop from `1` to `n` (inclusive)
Use a loop to iterate from `1` through `n`. Each number will be included in the product.

#### 3. Multiply `result` by each number in the loop
During each iteration, update `result` by multiplying it with the current number.

#### 4. After the loop ends, return `result`
Once the loop completes, `result` will contain the factorial value, which is returned.

---

### üßæ Example:
Let‚Äôs calculate `5!` step-by-step:

- Initialize: `result = 1`
- Loop begins:
  - `result = 1 * 1 = 1`
  - `result = 1 * 2 = 2`
  - `result = 2 * 3 = 6`
  - `result = 6 * 4 = 24`
  - `result = 24 * 5 = 120`
- Final result: `120`

‚úÖ So, `5! = 120`.


````python
def factorializeANumber(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Example usage:
print(factorializeANumber(5))  # Output: 120
````

### üß† Explanation of the Code

#### üîπ `def factorializeANumber(n):`
Defines a function named `factorializeANumber` that takes one parameter `n`, the number for which we want to compute the factorial.

---

#### üîπ `result = 1`
Initializes a variable `result` to store the running product of the numbers.  
We start with `1` because multiplying by `1` does not change the outcome and serves as the multiplicative identity.

---

#### üîπ `for i in range(1, n + 1):`
Loops from `1` to `n`, **inclusive**.  
The function `range(1, n + 1)` generates numbers from `1` to `n`.

Example:  
If `n = 5`, the loop variable `i` will take the values:  
`1 ‚Üí 2 ‚Üí 3 ‚Üí 4 ‚Üí 5`

---

#### üîπ `result *= i`
In each iteration, multiply `result` by the current value of `i`.  
This is shorthand for `result = result * i`.

---

### üîÑ Step-by-Step for `n = 5`:
- Initial: `result = 1`
- Multiply by 1 ‚Üí `1 * 1 = 1`
- Multiply by 2 ‚Üí `1 * 2 = 2`
- Multiply by 3 ‚Üí `2 * 3 = 6`
- Multiply by 4 ‚Üí `6 * 4 = 24`
- Multiply by 5 ‚Üí `24 * 5 = 120`

‚úÖ Final Result: `120`



### 6. `fancyRide(l, fares)`

#### Goal:
Choose the most luxurious ride that is affordable.

#### Steps:
1. You are given a list of fares per mile and a distance `l`.
2. Multiply each fare rate by the distance to get the total cost for each ride.
3. Compare each total cost to the maximum amount you can spend ($20).
4. Among the rides that are ‚â§ $20, pick the most expensive one.
5. Return the name of that ride.


# üéØ Goal
Choose the most luxurious ride (from a list of ride types) that costs $20 or less for a given distance l.

# üß† Step-by-Step Explanation
1. **Inputs:**
   You are given:
   - A float `l` representing the distance in miles.
   - A list of fares representing cost per mile for each ride type.
   
   The ride types are always listed in this fixed order:
   
   ```plaintext
   ["UberX", "UberXL", "UberPlus", "UberBlack", "UberSUV"]


````python
def fancyRide(l, fares):
    ride_types = ["UberX", "UberXL", "UberPlus", "UberBlack", "UberSUV"]
    max_affordable_index = 0

    for i in range(len(fares)):
        total_cost = fares[i] * l
        if total_cost <= 20:
            max_affordable_index = i  # Update to this more luxurious affordable ride

    return ride_types[max_affordable_index]

print(fancyRide(4.0, [0.6, 0.9, 1.5, 2.1, 2.6]))

````
