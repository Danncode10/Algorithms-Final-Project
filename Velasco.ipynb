# **CompanyBotStrategy**
---
## **TASK**
All four trainers have solved the task correctly, so the answer is (3 + 6 + 4 + 5) / 4 = 4.5.
Each CodeFights Company Bot is trained by engineers from that specific company. The way it works is that a representative group of engineers from each company is identified as trainers before the bot goes live, and they CodeFight against the bot during a training phase. The current training algorithm is definitely more complex, but let's assume it works this way:

For each trainer we collect two pieces of information per task [answerTime, correctness], where correctness is 1 if the answer was correct, -1 if the answer was wrong, and 0 if no answer was given. In this case, the bot's correct answer time for a given task would be the average of the answer times from the trainers who answered correctly. Given all of the training information for a specific task, calculate the bot's answer time.

**Example**
-   For

    trainingData = [ [3, 1],
                [6, 1],
                [4, 1],
                [5, 1]]
the output should be companyBotStrategy(trainingData) = 4.5.

-   For

    trainingData = [ [4, 1],
                [4, -1],
                [0, 0],
                [6, 1]]
the output should be companyBotStrategy(trainingData) = 5.0.

Only the 1st and the 4th trainers (1-based) submitted correct solutions, so the answer is (4 + 6) / 2 = 5.0.

-   For

    trainingData = [ [4, -1],
                [0, 0],
                [5, -1]]
the output should be companyBotStrategy(trainingData) = 0.0.

No correct answers were given for the task.

**Input/Output**

- **[execution time limit] 5 seconds (ts)**
- **[input] array.array.integer trainingData**
    The ith element of trainingData contains exactly 2 elements: trainingData[i][0] is the answer time for the ith person and trainingData[i][1] is correctness.

If a trainer answered correctly, then correctness equals 1;
If a trainer answered incorrectly, then correctness equals -1;
If a trainer didn't give any answer, answerTime and correctness both equal to 0.

*Guaranteed constraints:*

1 ≤ trainingData.length ≤ 20.

- **[output] float**

The time the bot will take to solve a specific task. Return 0 if none of the trainers answered correctly. Your output will be considered correct if its absolute error does not exceed 10-5 compared to our tests.

## **ANSWER**

```python
def companyBotStrategy(trainingData):
    total_time = 0
    correct_count = 0

    for answerTime, correctness in trainingData:
        if correctness == 1:
            total_time += answerTime
            correct_count += 1

    if correct_count == 0:
        return 0.0
    
    return total_time / correct_count
```

This challenge is about **analyzing performance data** from different trainers and computing an **average time** — but **only from the trainers who got the answer correct.**

Each entry in the **trainingData** list is a pair:

+ [answerTime, correctness]

+ answerTime: how long it took them to answer

+ correctness:

    + 1 means the answer was correct

    + -1 means the answer was incorrect

    + 0 means they didn’t answer at all

**The goal is to:**

+ **Only average the answer times of trainers who got the question correct.**

+ **If no one got it correct, return 0.0.**

**Implementation Key Points**
To solve the problem efficiently, here are the core steps:

Loop through the list of training data.

Check the correctness value:

If it is 1, the answer is correct.

Add the answer time to a total, and count the number of correct answers.

After the loop:

If no one was correct, return 0.0

Otherwise, divide the total time by the number of correct answers to get the average.

## **TASK**
Compare two integers given as strings.

Example

For a = "12" and b = "13", the output should be compareIntegers(a, b) = "less";

For a = "875" and b = "799", the output should be compareIntegers(a, b) = "greater";

For a = "1000" and b = "1000", the output should be compareIntegers(a, b) = "equal".

Hints

Input/Output

[execution time limit] 5 seconds (ts)

[input] string a A string representing a positive integer without leading zeroes.

Guaranteed constraints:

1 ≤ a.length ≤ 10.

[input] string b

A string representing a positive integer without leading zeroes.

Guaranteed constraints: 1 ≤ b.length ≤ 10

[output] string 'less' if int(a) < int(b), 'equal' if int(a) = int(b), and 'greater' if int(a) > int(b), where int(x) is equal to integer represented by the string x.
## **ANSWER**

```python
def compareIntegers(a: str, b: str) -> str:
    # Convert strings to integers
    int_a = int(a)
    int_b = int(b)

    # Compare the integers
    if int_a < int_b:
        return "less"
    elif int_a > int_b:
        return "greater"
    else:
        return "equal"

# Example usage:
print(compareIntegers("12", "13"))     # Output: "less"
print(compareIntegers("875", "799"))   # Output: "greater"
print(compareIntegers("1000", "1000")) # Output: "equal"
```

+ for "less"

+ > for "greater"

+ == for "equal"

**3.Return the Correct String:**

+ Based on the result of the comparison, return "less", "greater", or "equal".

# **ComposeRanges**
---
## **TASK**
Given a sorted integer array that does not contain any duplicates, return a summary of the number ranges it contains.

Example

For nums = [-1, 0, 1, 2, 6, 7, 9], the output should be composeRanges(nums) = ["-1->2", "6->7", "9"].
Hints

push()
Input/Output

[execution time limit] 5 seconds (ts)

[input] array.integer nums A sorted array of unique integers.

Guaranteed constraints:

0 ≤ nums.length ≤ 15, (231 - 1) ≤ nums[i] ≤ 231 - 1.

[output] array.string
## **ANSWER**

```python
def composeRanges(nums):
    if not nums:
        return []
    
    ranges = []
    start = nums[0]
    
    for i in range(1, len(nums)):
        # If the current number is not consecutive, we end the current range
        if nums[i] != nums[i - 1] + 1:
            end = nums[i - 1]
            if start == end:
                ranges.append(f"{start}")
            else:
                ranges.append(f"{start}->{end}")
            start = nums[i]  # Start a new range
    
    # Add the last range
    if start == nums[-1]:
        ranges.append(f"{start}")
    else:
        ranges.append(f"{start}->{nums[-1]}")
    
    return ranges

# Example usage
nums = [-1, 0, 1, 2, 6, 7, 9]
print(composeRanges(nums))  # Output: ["-1->2", "6->7", "9"]
```

In this problem, we are given a **sorted list of numbers** that don’t repeat. Our goal is to **group together numbers that appear in a continuous sequence** (like 1, 2, 3) and represent them as a **range** (like "1->3"). If a number **stands alone** and isn’t part of a sequence, we just return it as it is.

For example:

+ Given: [-1, 0, 1, 2, 6, 7, 9]

+ Output: ["-1->2", "6->7", "9"]

We're summarizing the list in a cleaner format that shows where the numbers are continuous (like a chain) and where they are not.

**Here’s what our code needs to handle:**

**1.Look at each number in the list one by one.**

**2.Check if the next number continues the sequence (i.e., is one greater than the current).**

**3.Keep track of the start and end of the current range.**

**4.When the sequence breaks (the next number skips), we:**

+ Save the range like "start->end" if it has multiple numbers.

+ Or just the number "start" if it's alone.

**5.Repeat this until all numbers are checked.**

**6.Finally, return a list of strings summarizing the ranges.**

## **💡 Key Implementation Points**
+ **Start and End Pointers:** We’ll use these to track where a range begins and ends.

+ **Loop:** Go through the list using a loop.

+ **Condition:** Check if the current number +1 equals the next — if so, it's part of the same range.

+ **Edge Cases:** Handle single numbers and the end of the list carefully.

+ **Result Storage:** Use .append() (or push() in other languages) to store formatted strings into a result list.

# **ConfirmEnding**
---
## **TASK**
Check if a string (first argument, str) ends with the given target string (second argument, target).

This challenge can be solved with the .endsWith() method, which was introduced in ES2015. But for the purpose of this challenge, we would like you to use one of the JavaScript substring methods instead.

Remember to use Read-Search-Ask if you get stuck. Write your own code.

Example

confirmEnding("Abstraction", "action") returns true;
confirmEnding("Open sesame", "pen") returns false;
Hints

substr()
## **ANSWER**

```python
def confirm_ending(string, target):
    # Get the length of the target
    target_length = len(target)
    # Slice the end of the string to compare with target
    return string[-target_length:] == target

# Example usage:
print(confirm_ending("Abstraction", "action"))  
print(confirm_ending("Open sesame", "pen"))     
print(confirm_ending("Hello world", "world")) 
print(confirm_ending("Python", "thon"))       
```

# **ConvertCelsiusToFahrenheit**
---
## **TASK**
The algorithm to convert from Celsius to Fahrenheit is the temperature in Celsius times 9/5, plus 32.

You are given a variable celsius representing a temperature in Celsius. Use the variable fahrenheit already defined and assign it the Fahrenheit temperature equivalent to the given Celsius temperature. Use the algorithm mentioned above to help convert the Celsius temperature to Fahrenheit.

Example

celsiusToFahrenheit(-30) returns -22;
celsiusToFahrenheit(-10) returns 14;
celsiusToFahrenheit(0) returns 32;
## **ANSWER**

```python
def celsiusToFahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Example usage
print(celsiusToFahrenheit(-30))  # Output: -22.0
print(celsiusToFahrenheit(-10))  # Output: 14.0
print(celsiusToFahrenheit(0))    # Output: 32.0
```

This code is designed to **convert temperatures from Celsius to Fahrenheit.** That means if you input a temperature like 0°C (which is freezing point in Celsius), the function will give you the equivalent in Fahrenheit, which is 32°F.

We expect it to:

+ Accept a number (in Celsius).

+ Do the math to change that number into Fahrenheit.
Give back the result (the converted temperature).

🧠 **How Does It Work (Key Implementation Points)?**
**1.Function Name**

+ celsiusToFahrenheit(celsius) is the function’s name.

+ The word "celsius" inside the parentheses is the input — this is where you put your temperature.

**2.The Formula**

+ Inside the function, the formula used is:

        fahrenheit=(celsius * 9/5) + 32
+ This is the standard formula to convert Celsius into Fahrenheit.

**3.Return Statement**
+ The function gives back the converted value using return fahrenheit.

**4.Example Usage**

+ The lines like print(celsiusToFahrenheit(-30)) test the function using sample values.

+ It shows how cold -30°C becomes -22°F in the Fahrenheit system.

# **ConvertString**
---
## **TASK**
You are given the string s. Your friend just asked you if it's possible to change the string from s to a string t by removing some characters from it. You're a pro at programming, so you decided to create a program to determine this.

Example

For s = "ceoydefthf5iyg5h5yts" and t = "codefights", the output should be convertString(s, t) = true.

For s = "addbyca" and t = "abcd", the output should be convertString(s, t) = false.

Hints

concat()
split()
Input/Output

[execution time limit] 4 seconds (js)

[input] string s A string with alphanumeric characters.

Guaranteed constraints:

1 ≤ s.length ≤ 1000.

[input] string t

A string with alphanumeric characters.

Guaranteed constraints: 1 ≤ t.length ≤ 1000.

[output] boolean Return true if it is possible to convert s to t, otherwise return false.
## **ANSWER**

```python
def convertString(s, t):
    i, j = 0, 0  # i for s, j for t
    
    # Traverse both strings
    while i < len(s) and j < len(t):
        if s[i] == t[j]:
            j += 1  # Move pointer for t when characters match
        i += 1  # Always move pointer for s
    
    # If we have traversed all of t, it means we can form t from s
    return j == len(t)

# Test cases
print(convertString("ceoydefthf5iyg5h5yts", "codefights"))  # True
print(convertString("addbyca", "abcd"))  # False
```

to determine if you can transform the given string s into another string t by **removing some characters** without changing their order.

Think of it like this: Imagine s is a long text, and t is a word you're trying to find within s. If the letters of t appear in order within s (even with gaps between them), then the transformation is possible.

**For example:**

s = "ceoydefthf5iyg5h5yts", t = "codefights" → We can remove extra characters and still find "codefights" in order, so return true.

s = "addbyca", t = "abcd" → The order of "abcd" isn't preserved within "addbyca", so return false.

**.Key Implementation Points**
To solve this in a structured way:

**1.Track the order of letters in t**

**2.Use a pointer or index approach**
